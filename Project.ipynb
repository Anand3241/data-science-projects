```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import statsmodels.api as sm
import matplotlib.dates as mdates
from pandas import DataFrame,Series
from pylab import rcParams
from numpy.random import randn
from matplotlib import pyplot
```

```python
df=pd.read_csv('MARUTI_data.csv',index_col='Date',parse_dates=True)
df.head()
```

# Visualize  the Data set


```python
rcParams['figure.figsize']=20,5
sns.set_style('whitegrid')
```

```python

df['volume'].plot()
```

```python
df2=df.sample(n=100,random_state=25,axis=0)
plt.xlabel('Volume Date')
plt.ylabel('Market Volume Data based on the Year')
plt.title('Market Volume Vs Year')
df2['volume'].plot()
```

# Visualize the Opening of the share based on the Year

```python
df2=df.sample(n=100,random_state=25,axis=0)
plt.xlabel('Open Date')
plt.ylabel('Stock open  based on the Year')
plt.title('Market open Vs Year')
df2['open'].plot()
```

# Visualize the Open vs High Vs Low Vs Close of the share bases on the year

```python
df2[['open','high','low','close']].plot.bar(stacked=True)
```

```python
df2[['open','high','low','close']].plot.area(alpha=0.4)
```

```python

df2[['open','high','low','close']].plot.box()
```
